## Importing Library

```python
import geopandas as gpd
```

## Dataset used [Geospatial Learn course data](https://www.kaggle.com/alexisbcook/geospatial-learn-course-data)

```python
full_data = gpd.read_file("DEC_lands/DEC_lands.shp")
full_data.head()
```

## Using only the subset of column

```python
data = full_data.loc[:, ["CLASS", "COUNTY", "geometry"]].copy()
```

## How many lands of each type are there?

```python
data.CLASS.value_counts()
```

## Select lands that fall under the "WILD FOREST" or "WILDERNESS" category

```python
wild_lands = data.loc[data.CLASS.isin(['WILD FOREST', 'WILDERNESS'])].copy()
wild_lands.head()
```

# Let's plot the Map

```python
wild_lands.plot()
```

## View the first five entries in the "geometry" column

```python
wild_lands.geometry.head()
```

## we will create three more GeoDataFrames, containing campsite locations (Point), foot trails (LineString), and county boundaries (Polygon).

```python
# Campsites in New York state (Point)
POI_data = gpd.read_file("DEC_pointsinterest/DEC_pointsinterest/Decptsofinterest.shp")
campsites = POI_data.loc[POI_data.ASSET=='PRIMITIVE CAMPSITE'].copy()

# Foot trails in New York state (LineString)
roads_trails = gpd.read_file("DEC_roadstrails/DEC_roadstrails/Decroadstrails.shp")
trails = roads_trails.loc[roads_trails.ASSET=='FOOT TRAIL'].copy()

# County boundaries in New York state (Polygon)
counties = gpd.read_file("NY_county_boundaries/NY_county_boundaries/NY_county_boundaries.shp")
```

## Define a base map with county boundaries and Add wild lands, campsites, and foot trails to the base map

```python
ax = counties.plot(figsize=(10,10), color='none', edgecolor='gainsboro', zorder=3)
wild_lands.plot(color='lightgreen', ax=ax)
campsites.plot(color='maroon', markersize=2, ax=ax)
t = trails.plot(color='yellow', markersize=1, ax=ax)
```
